Function that adds two numbers and determines whether the result is odd or even:

```python
def evenOrOdd(num1, num2):
  num1 = int(num1)
  num2 = int(num2)
  sum = num1 + num2
  if sum % 2 == 0:
    print(f'{sum} is even')
  else:
    print(f'{sum} is odd')
```

```python
evenOrOdd(-43, 81)
```

#### Convert Fahrenheit to Celsius:
######Celsius= (Fahrenheit - 32) * 5/9

```python
def temp_convert(fahrenheit):
  celsius = (fahrenheit - 32) * 5/9
  return round(celsius,3)
```

```python
temp_convert(63)
```

Leap Year determination:

"A leap year is a year that is divisible by 4 except for years that are evenly divisible by 100, although years divisible by 400 are leap years".

```python
def LeapYeaorNae(year):
  if (year % 4 == 0 and year % 100 !=0) or (year % 400 == 0):
    print('Its a Leap Year!')
  else:
    print('Not a Leap Year')
```

```python
LeapYeaorNae(2032)
```

Senator Tweets downloaded from [here](https://btech.instructure.com/courses/591588/files/113256798?wrap=1).

```python
import pandas as pd

df = pd.read_csv('/content/senator_tweets.csv')
```

```python
df.head()
```

```python
#New Character Count Column added
df['Character_Count'] = df['text'].str.len()
```

```python
df.head()
```

```python
# Correlation coefficient between the length of a tweet and the number of replies that it gets
round(df['Character_Count'].corr(df['replies']),6)
```

```python
# Shortest recorded tweet in data set:
df[df['Character_Count'] == df['Character_Count'].min()]
```

```python
# Filter showing only tweets mentioning "immigration".
df.loc[df['text'].str.contains('immigration')]
```

```python
# Filter showing tweets that contain "Canada"
df.loc[df['text'].str.contains('Canada')]
```

```python
# Tweets of 2010 Filter
df.loc[df['created_at'].str.contains('2010')]
```

```python
# Month column created
df['Month'] = df['created_at'].str.split("/").str[0]
```

```python
df.head(100)
```

```python
#Number of tweets by month
monthly = df['Month'].value_counts()
```

```python
# Word Count column created
df['Words'] = df['text'].str.split().str.len()
```

```python
df.head()
```

```python
# Most word count per state
avg_per_state = df.groupby('state')['Words'].mean()
avg_per_state.sort_values(ascending=False)
```

```python
import pandas as pd

df = pd.read_csv('/content/invoice_lines.csv')
```

```python
df.head()
```

```python
#Colors from description column used to create Color column.
df['color'] = df['Description'].str.extract('\(([A-Za-z]+)\)')[0]
```

```python
df.head()
```

```python
#how many by color
df.groupby('color')['Quantity'].sum().sort_values(ascending=False)
```

```python
#filter for contains "slipper"
slipper = df[df['Description'].str.contains('slipper')]
```

```python
slipper
```

```python
slipper.loc[:,'size'] = slipper['Description'].str.extract('\s([A-Z]+)$')[0]
```

```python
slipper.groupby('size')['Quantity'].sum()
```
