```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
```

```python
# Question 1 & 2: Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state.
# Which state are most of the companies from? California (CA)
plt.figure(figsize = (9,5))
sns.countplot(y='state', data=df)
plt.title('Company Count by State')
plt.xticks(rotation = 90)
plt.show()

# Sorted graph
df_sorted = df['state'].value_counts().sort_values(ascending=False)

plt.figure(figsize = (9,5))
sns.countplot(y='state', data = df, order = df_sorted.index, color = 'purple')
plt.title('Company Count by State')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Question 3 & 4: Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.
# Which industry has the most job openings?  Fintech
sns.barplot(x='Industry', y='job_openings', data=df, estimator = sum, errorbar = None)
plt.xticks(rotation = 45)
plt.title('Job Openings by Industry')
plt.show()
```

```python
#Question 5: Using matplotlib/seaborn, create a line chart that shows the average growth rate for each founded year.

# Create aggregated df
df_rate = df.groupby('founded')['growth_percentage'].mean().reset_index()

# Graph 
sns.lineplot(x='founded', y='growth_percentage', data=df_rate)
plt.title('Average Growth Rate per Year Founded')
plt.show()

# vs unagreggated data graph: 
sns.lineplot(x='founded', y='growth_percentage', data=df)
```

```python
# Question 6: From 2000 to 2020 the growth percentage rate appears to be increasing.
```

```python
# Question 7: Using matplotlib/seaborn, create a scatter plot that compares the number of employees to estimated revenues for each company.
sns.scatterplot(x='employees', y='estimated_revenues', data=df, color='green')
plt.title('Number of Employees vs Estimated Revenue per Company')
plt.ylabel('Estimated Revenues(millions)')
plt.xlabel('Employees')
plt.show()
```

```python
# Question 9: create a histogram that measures the number of companies founded in each year.
plt.figure(figsize=(6, 3))
sns.histplot(x='founded', data=df, bins = 20)
plt.title('Number of Companies Founded Each Year')
plt.xlabel('Year Founded')
plt.show()
```
